```python
#Chainage avant 1
import time
def chainage_avant(faits, regles):
    trace = []
    regles_appliquees = set()
    nouvelles_conclusions = True
    while nouvelles_conclusions:
        nouvelles_conclusions = False
        for regle in regles:
            if regle['nom'] in regles_appliquees:
                continue
            if all(faits.get(cle) == valeur for cle,valeur in regle["conditions"].items()):
                nouvelle = False
                for cle,valeur in regle["conclusion"].items():
                    if faits.get(cle) != valeur:
                        faits[cle] = valeur
                        nouvelle = True
                if nouvelle:
                    trace.append(f"Règle {regle['nom']} appliquée : {regle['conclusion']}")
                    nouvelles_conclusions = True
                    regles_appliquees.add(regle["nom"])
    return faits, trace
```

```python
#Tkinter interface
def lancer_diagnostic():
    faits = {
        "plante": entree_plante.get().strip().lower(),
        "saison": entree_saison.get().strip().lower(),
        "humidite_ambiante": entree_humidite.get().strip().lower(),
        "sol": entree_sol.get().strip().lower(),
        "lumiere": entree_lumiere.get().strip().lower(),
        "temperature": entree_temp.get().strip().lower(),
        "besoin_eau": entree_eau.get().strip().lower(),
        "etat_feuilles": entree_feuilles.get().strip().lower(),
        "etat_sol": entree_sol_etat.get().strip().lower()
    }
    
    resultats, trace = chainage_avant(faits.copy(), regles)
    
    texte_resultat.delete("1.0", tk.END)
    texte_resultat.insert(tk.END, "== Règles Appliquées ==\n")
    for ligne in trace:
        texte_resultat.insert(tk.END, ligne + "\n")
        
    texte_resultat.insert(tk.END, "\n== Conclusions ==\n")
    for cle,valeur in resultats.items():
        if cle not in ["plante", "saison", "humidite_ambiante", "sol", "lumiere", "temperature", "besoin_eau", "etat_feuilles", "etat_sol"]:
            texte_resultat.insert(tk.END, f"{cle} : {valeur}`\n")
```

```python
#Fenêtre principale
fenetre = tk.Tk()
fenetre.title("Smart Pot - Système Expert")

#Champs de saisie
etiquettes = [
    "Plante", "Saison", "Humidité ambiante", "Sol", "Lumière", "Température", "Besoin en eau", "Etat des feuilles", "Etat du sol"
]

entrees = []

for i,texte in enumerate(etiquettes):
    label = tk.Label(fenetre, text=texte)
    label.grid(row=i, column=0, padx=5, pady=2)
    entree = tk.Entry(fenetre, width=30)
    entree.grid(row=i, column=1, padx=5, pady=2)
    entrees.append(entree)

(entree_plante, entree_saison, entree_humidite, entree_sol, entree_lumiere, entree_temp, entree_eau, entree_feuilles, entree_sol_etat) = entrees


#Bouton diagnostic
btn = tk.Button(fenetre, text="Lancer le diagnostic", command=lancer_diagnostic)
btn.grid(row=len(etiquettes), columnspan=2, pady=10)

#Zone d'affichage des résultats
texte_resultat = tk.Text(fenetre, width=70, height=15)
texte_resultat.grid(row=len(etiquettes)+1, columnspan=2, padx=5, pady=5)

#Lancer l'application
fenetre.mainloop()
```
